```python
dataset = pd.read_table('abalone.data', sep=',', header=None, names=['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings'])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables = dataset.iloc[:,1:8]
vif = pd.DataFrame()
```

```python
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
```
