```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats
```

```python
credit = pd.read_csv('creditpromo.csv')
```

```python
credit.head()
```

```python
standard = credit[credit['insert']=='Standard']
new = credit[credit['insert']=='New Promotion']
```

```python
standard
```

```python
new
```

```python
stats.ttest_ind(standard['dollars'], new['dollars'])
```

```python
sm.stats.ttest_ind(standard['dollars'], new['dollars'], alternative='two-sided', usevar='pooled', value=0)
```

```python
sm.stats.ttest_ind(standard['dollars'], new['dollars'], alternative='two-sided', usevar='unequal', value=0)
```
